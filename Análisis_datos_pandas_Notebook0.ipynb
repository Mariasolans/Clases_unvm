## 1. Principales métodos de Pandas
**[Pandas](http://pandas.pydata.org)** es una librería de programación en Python utilizada principalmente para la manipulación y análisis de datos. Fue creada por Wes McKinney y se ha convertido en una herramienta fundamental en el mundo de la ciencia de datos y análisis de datos en Python. Algunas de las características clave de Pandas incluyen:

1. **Estructuras de Datos**: Pandas proporciona dos estructuras de datos principales: **Series** y **DataFrame**.

   - Una **Serie** es una estructura unidimensional similar a un array o una lista, pero con etiquetas en lugar de índices numéricos.
   
   - Un **DataFrame** es una estructura bidimensional similar a una tabla de base de datos o una hoja de cálculo de Excel. Contiene filas y columnas, y cada columna puede contener diferentes tipos de datos.

2. **Lectura y Escritura de Datos**: Pandas ofrece herramientas para leer y escribir datos en una variedad de formatos, incluyendo CSV, Excel, SQL, JSON, y más. Esto facilita la importación y exportación de datos desde y hacia diferentes fuentes.

3. **Manipulación de Datos**: Pandas permite realizar operaciones como filtrado, selección, agrupación, agregación, fusión y transformación de datos de manera eficiente. Esto es fundamental para el análisis de datos y la limpieza de datos.

4. **Manejo de Valores Faltantes**: Pandas ofrece métodos para trabajar con valores faltantes (NaN) de manera efectiva, como la eliminación de filas o columnas con valores faltantes o la imputación de valores faltantes.

5. **Indexación y Etiquetado**: Pandas permite etiquetar y acceder a datos mediante índices y columnas, lo que facilita la manipulación de datos basada en etiquetas en lugar de índices numéricos.

6. **Visualización de Datos**: Aunque Pandas en sí no es una herramienta de visualización, se integra fácilmente con bibliotecas de visualización como Matplotlib y Seaborn para crear gráficos y visualizaciones a partir de los datos en un DataFrame.

7. **Eficiencia y Rendimiento**: Pandas está diseñado para trabajar eficientemente con conjuntos de datos grandes y complejos, lo que lo hace adecuado para aplicaciones de análisis de datos de la vida real.


**[Numpy](https://numpy.org/)** es una biblioteca fundamental en el ecosistema de Python que proporciona soporte para trabajar con arreglos multidimensionales y matrices, así como funciones matemáticas de alto rendimiento para realizar operaciones en estos arreglos. Fue creado para facilitar la computación numérica y científica en Python y es ampliamente utilizado en campos como la ciencia de datos, la ingeniería, la física, la bioinformática y muchas otras disciplinas.

Aquí hay algunas características clave de NumPy:

1. **Arreglos NumPy**: El elemento central de NumPy es el objeto `ndarray`, que es un arreglo multidimensional que permite almacenar datos de manera eficiente en memoria. Estos arreglos pueden tener una o más dimensiones y contener elementos del mismo tipo de datos.

2. **Operaciones Vectorizadas**: NumPy permite realizar operaciones matemáticas y lógicas en arreglos de manera eficiente y vectorizada, lo que significa que puedes realizar operaciones en arreglos completos en lugar de iterar sobre elementos individuales. Esto mejora significativamente el rendimiento en comparación con las listas de Python estándar.

3. **Funciones Matemáticas**: NumPy proporciona una amplia gama de funciones matemáticas de alto rendimiento, como trigonometría, álgebra lineal, estadísticas, generación de números aleatorios y más.

4. **Indexación y Rebanado Avanzados**: Puedes acceder y modificar elementos específicos de un arreglo NumPy utilizando una variedad de técnicas de indexación y rebanado (slicing), lo que facilita la manipulación de datos en arreglos.

5. **Integración con Otras Bibliotecas**: NumPy se utiliza comúnmente en combinación con otras bibliotecas de Python, como SciPy (para computación científica avanzada), Matplotlib (para visualización de datos), y pandas (para análisis de datos), lo que permite construir aplicaciones y flujos de trabajo de análisis de datos completos.

6. **Eficiencia y Rendimiento**: NumPy está escrito en C y optimizado para un rendimiento óptimo, lo que lo hace adecuado para el procesamiento de datos grandes y complejos.

### Agrupamiento

En  general, agrupar datos en Pandas funciona como sigue:


1. Primero, el método `groupby` divide las `grouping_columns` por sus valores. Estos devienen los nuevos índices del dataframe generado.
2. Luego se seleccionan las columnas de interés  (`columns_to_show`). Si `columns_to_show` no esta incluido, todas las columnas serán incluidas.
3. Finalmente, uno  o mas funciones se aplican a los grupos obtenidos seleccionando columnas.

Por ejemplo, podemos responder **¿Cómo impacta la clase en los niveles de satisfacción?**

```python
# Calcular la distribución de las clases por satisfacción y convertirla en porcentajes
class_distribution = df.groupby('Class')['satisfaction'].value_counts(normalize=True).unstack().mul(100)
class_distribution
```

```python


# Crear un gráfico de barras para cada fila en el DataFrame "class_distribution"
plt.figure(figsize=(12, 6))
ax=class_distribution.plot(kind='bar', stacked=False, colormap='Set3')

# Personalizar el gráfico
plt.title('Distribución de Satisfacción por Clase')
plt.xlabel('Clase')
plt.ylabel('Porcentaje')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para una mejor legibilidad

# Agregar porcentajes encima de las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')
# Mostrar el gráfico
plt.legend(title='Satisfacción', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Tablas de resumen

Suponga que queremos ver como se distribuyen las observaciones de nuestro dataframe  en el contexto de dos variables  the context of two variables - `Type of Travel` y `satisfaction`. Para hacerlo armamos una **tabla de contingencia** usindo el método `crosstab`:


```python
cross_tab=pd.crosstab(df['Type of Travel'], df['satisfaction'])
# Convertir los conteos a porcentajes
percentage_cross_tab = cross_tab.apply(lambda row: row / row.sum() * 100, axis=1)
print(percentage_cross_tab)
```

```python


# Crear un gráfico de barras apiladas a partir de percentage_cross_tab
fig, ax = plt.subplots(figsize=(12, 8))  # Ajusta el tamaño de la figura

percentage_cross_tab.plot(kind='bar', stacked=True, colormap='Set3', ax=ax)

# Personalizar el gráfico
plt.title('Distribución de Satisfacción por Tipo de Viaje')
plt.xlabel('Tipo de Viaje')
plt.ylabel('Porcentaje')
plt.xticks(rotation=0)  # No rotar las etiquetas del eje x

# Obtener el primer valor (porcentaje) dentro de la primera barra
first_bar = ax.patches[0]
first_value = first_bar.get_height()

# Agregar porcentajes dentro de las barras
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        if p == first_bar:
            ax.annotate(f'{first_value:.2f}%', (p.get_x() + p.get_width() / 2., height / 2),
                        ha='center', va='center', fontsize=10, color='black', xytext=(0, -20),
                        textcoords='offset points')
        else:
            ax.annotate(f'{height:.2f}%', (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                        textcoords='offset points')

# Mostrar el gráfico
plt.legend(title='Satisfacción', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Las tablas pivote son tablas dinámicas son una herramienta poderosa en el análisis de datos que permiten resumir y analizar datos tabulares de manera eficiente. Para implementar estas tablas se usa el método `pivot_table`con los parámetros:

- `data`: El DataFrame original que contiene los datos que deseas resumir.
- `values`: Las columnas que deseas agregar o resumir en la tabla dinámica.
- `index`: Las columnas que se utilizarán como índices en la tabla dinámica.
- `columns`: Las columnas que se utilizarán para crear las columnas en la tabla dinámica.
- `aggfunc`: La función de agregación que se utilizará para resumir los datos (por ejemplo, suma, promedio, conteo, etc.).



```python

df.pivot_table( ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes'], ['satisfaction'], aggfunc='mean')
```
